```python
import tensorflow as tf
import os
```

```python
BASE_DIR = os.getcwd()  # backend/
H5_DIR = os.path.join(BASE_DIR, "saved_models")
TFLITE_DIR = os.path.join(BASE_DIR, "saved_models", "tflite_models")

os.makedirs(TFLITE_DIR, exist_ok=True)

print("H5 DIR:", H5_DIR)
print("TFLITE DIR:", TFLITE_DIR)
```

```python
def convert_h5_to_tflite(h5_path, tflite_path):
    print(f"\nðŸ”„ Loading model: {os.path.basename(h5_path)}")

    model = tf.keras.models.load_model(h5_path)

    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]

    tflite_model = converter.convert()

    with open(tflite_path, "wb") as f:
        f.write(tflite_model)

    h5_size = os.path.getsize(h5_path) / (1024 * 1024)
    tflite_size = os.path.getsize(tflite_path) / (1024 * 1024)

    print(f"âœ… Saved: {os.path.basename(tflite_path)}")
    print(f"ðŸ“‰ Size: {h5_size:.2f} MB â†’ {tflite_size:.2f} MB")
```

```python
for file in os.listdir(H5_DIR):
    if file.endswith(".h5"):
        h5_path = os.path.join(H5_DIR, file)
        tflite_path = os.path.join(
            TFLITE_DIR,
            file.replace(".h5", ".tflite")
        )

        convert_h5_to_tflite(h5_path, tflite_path)
```
